```python
import quimb as qu
import matplotlib.pyplot as plt
import numpy as np
```

```python
L=int(input('Enter no. of Particles:'))
```

```python
TM=0
for i in range(L):
    TM=TM+(qu.ikron(qu.spin_operator('x'),[2]*L,inds=[i]))
    
```

```python
H1=qu.ham_heis(L,j=(-4,0,0),b=(0.001,0,-1),sparse=True)
H2=qu.ham_heis(L,j=(-4,0,0),b=(-0.2,0,-0.5),sparse=True)
```

```python
S_i=qu.qu(qu.groundstate(H1),qtype='ket')
```

```python
el=list()
evo=qu.Evolution(S_i,H2)
for pt in evo.at_times(np.arange(0,30,0.1)):
    e=qu.expec(pt,TM)
    el.append(2*e/L)
    
```

```python
plt.plot(np.arange(0,30,0.1),el)
```

```python
gl=list()
for i in np.arange(0,1,0.01):
    H=qu.ham_heis(L,j=(-1,0,0),b=(0.001,0,i),sparse=True)
    g=qu.groundstate(H)
    ge=qu.expec(g,TM)
    gl.append(2*ge/L)
plt.plot(np.arange(0,1,0.01),gl)    
```

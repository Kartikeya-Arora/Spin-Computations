```python
from numpy import *
import random
from matplotlib.pyplot import* 

hbar=1
X =array([[0,1],[1,0]],dtype=complex128)
Y =array([[0,0-1.j],[0+1.j,0]],dtype=complex128)
Z =array([[1,0],[0,-1]],dtype=complex128)
print(X)
print()
print(Y)
print()
print(Z)
```

```python
x = random.random()
y = random.random()
z = random.random()
    
```

```python
H=(x*X)+(y*Y)+(z*Z)
print(H)
```

```python
v,w=linalg.eigh(H)
print(v)
print(w)
```

```python
p=array(w[:,0]) 
psi=p.reshape(2,1)
print('psi=',psi)
hpsi=p.conj()
print('hpsi=',hpsi)
```

```python
M1=dot(hpsi,dot(X,psi))
M2=dot(hpsi,dot(Y,psi))
M3=dot(hpsi,dot(Z,psi))





```

```python
print('Mx=',M1)
print('My=',M2)
print('Mz=',M3)
```

```python
x = random.random()
y = random.random()
z = random.random()

H=(x*X)+(y*Y)+(z*Z)
v,P=linalg.eigh(H)
Pinv=transpose(conj(P))


print('v=',v)
print('P=',P)
print('Pinv=',Pinv)

p=array(P[:,0]) 
psi0=p.reshape(2,1)
print('psi0=',psi0)

D=(diag(v))


print('D=',D)
```

```python
time=arange(1,11,0.1)
list_psit=list()
m1t=list()
m2t=list()
m3t=list()
for t in time:
    D=diag(exp((((-v)*t)/hbar)*1.j))

    
    U=dot(P,dot(D,Pinv))
    psit=dot(U,psi0)
    list_psit.append(psit)
    
    hp=psit.reshape(1,2)
    hpsit=hp.conj()
    M1t=dot(hpsit,dot(X,psit))
    M2t=dot(hpsit,dot(Y,psit))
    M3t=dot(hpsit,dot(Z,psit))
    m1t.append(M1t[0])
    m2t.append(M2t[0])
    m3t.append(M3t[0])    
    
    D=diag(v)
    
   
  
```

```python
## for Mxt
plot(time,m1t)
xlabel('time')
ylabel('Mxt')
```

```python
## for Myt
plot(time,m2t)
xlabel('time')
ylabel('Myt')
```

```python
## for Mzt
plot(time,m3t)
xlabel('time')
ylabel('Mzt')
```
